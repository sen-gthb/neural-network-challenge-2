## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# check for any missing values in the data
attrition_df.isnull().sum()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[["Attrition", "Department"]]

y_df.head()

```

```python
#find out how many unique values are in the Attrition column and the Department columns.

print(y_df['Attrition'].value_counts())
print(y_df['Department'].value_counts())
```

```python
# Create a list of at least 10 column names to use as X data
selected_columns = [
    "Age",
    "NumCompaniesWorked",
    "DistanceFromHome",
    "TrainingTimesLastYear",
    "HourlyRate",
    "YearsSinceLastPromotion",
    "JobSatisfaction",
    "TotalWorkingYears",
    "YearsAtCompany",
    "PercentSalaryHike",
    "YearsInCurrentRole"]


# Create X_df using your selected columns
X_df = attrition_df[selected_columns]



# Show the data types for X_df and the shape
print(X_df.dtypes)
print(X_df.shape)
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=78)


```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary

# NOT NEEDED FOR THE FEATURES I CHOSE BUT IF I DID I WOULD USE:
# X_train = X_train.apply(pd.to_numeric)
# X_test = X_test.apply(pd.to_numeric)

```

```python
# Create a StandardScaler
from sklearn.preprocessing import StandardScaler


# Fit the StandardScaler to the training data
X_scaler = StandardScaler().fit(X_train)


# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import OneHotEncoder

# Fit the encoder to the training data
enc = OneHotEncoder(sparse_output=False, handle_unknown="ignore", drop=None)
enc.fit(np.array(y_train["Department"]).reshape(-1, 1))

# Create two new variables by applying the encoder
# to the training and testing data

y_train_department = enc.transform(np.array(y_train["Department"]).reshape(-1, 1))
y_test_department = enc.transform(np.array(y_test["Department"]).reshape(-1, 1))

print(y_train_department)
print(y_test_department)
```

```python
# Create a OneHotEncoder for the Attrition column
enc = OneHotEncoder(sparse_output=False)


# Fit the encoder to the training data
enc.fit(np.array(y_train["Attrition"]).reshape(-1, 1))


# Create two new variables by applying the encoder
# to the training and testing data
y_train_attrition = enc.transform(np.array(y_train["Attrition"]).reshape(-1, 1))
y_test_attrition = enc.transform(np.array(y_test["Attrition"]).reshape(-1, 1))

print(y_train_attrition)
print(y_test_attrition)

```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
input_features = len(X_train_scaled[0])


# Create the input layer
input_features = len(X_train_scaled[0])
inputs = layers.Input(shape=input_features,name="inputs")


# Create at least two shared layers
shared_1 = layers.Dense(units=64, activation='relu', name="shared_1")(inputs)
shared_2 = layers.Dense(units=128, activation='relu', name="shared_2")(shared_1)
```

```python
print(inputs)
print(shared_1)
print(shared_2)
```

```python
y_train_department.shape[1]
```

```python
# Create a branch for Department with a hidden layer and an output layer

department_hidden = layers.Dense(units=32, activation='softmax', name='department_hidden')(shared_2)
department_output = layers.Dense(units=3, activation='softmax', name='department_output')(department_hidden)


```

```python
y_train_attrition.shape[1]
```

```python
# Create a branch for Attrition with a hidden layer and an output layer

attrition_hidden = layers.Dense(units=32, activation='softmax', name='attrition_hidden')(shared_2)
attrition_output = layers.Dense(units=2, activation='softmax', name='attrition_output')(attrition_hidden)




```

```python
# Create the model
model = Model(inputs=inputs, outputs=[department_output, attrition_output])

# Compile the model
model.compile(optimizer='adam',
              loss={'department_output': 'categorical_crossentropy', 'attrition_output': 'categorical_crossentropy'},
              metrics={"department_output":'accuracy', "attrition_output":'accuracy'})

# Summarize the model
model.summary()


```

```python
# Train the model
model.fit(X_train_scaled, {'department_output': y_train_department, 'attrition_output': y_train_attrition}, epochs=100, shuffle=True, verbose=2)



```

```python


# Evaluate the model with the testing data
evaluation_results = model.evaluate(X_test_scaled, {'department_output': y_test_department, 'attrition_output': y_test_attrition}, verbose=2)

# Print the evaluation results
print("Evaluation results:", evaluation_results)

```

```python
# Print the accuracy for both department and attrition
print(f"Department Accuracy: {evaluation_results[3]*100}")
print(f"Attrition Accuracy: {evaluation_results[4]*100}")
```

Because this dataset is imbalanced, meaning that one class in each y feature (no attrition; Research & Development) is much more prevalent than the other (attrition; Sales, Human Resources), accuracy alone can be misleading. A model that simply predicts the majority class for all instances could achieve high accuracy, but it would not be useful in practice.
We would also want to look at the business context. For example, if the primary goal were to identify employees at risk of leaving in order to intervene and retain them, a model with high recall (minimizing false negatives) would be more valuable, even if it sacrifices some precision.
2. I started with the reLu activation for the shared and hidden layers with Softmax only in the output layer of the y data, because ReLU is a common choice for activation functions in the input and hidden layers of neural networks due to its simplicity, efficiency, and effectiveness in learning complex representations from data. However, the model wasn't performing super well so I decided to retain reLu in the shared layers, but try Softmax across the hidden layers in addition to the output layers. Softmax activation is primarily used in the output layer of neural networks for multi-class classification tasks. It's particularly useful when you have more than two classes in your target variable. In this case, it helped the model to perform even better when it was used across the hidden and output layers.
